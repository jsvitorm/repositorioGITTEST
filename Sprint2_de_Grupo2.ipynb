CRIAMOS UM DF COM AS COLUNAS CIE (IDENTIFICADOR DO USUARIO ESCOLA), CNPJ (IDENTIFICADOR DO USUARIO FORNECEDOR) E AVALIAÇÃO

(OS DADOS VAZIOS ESTAMOS PREENCHENDO COM A MÉDIA DE AVALIAÇÃO DAQUELE FORNECEDOR)

ESTAMOS AGRUPANDO POR CIE E CNPJ

```python
df_dadosneed
df_agrupado = df_dadosneed.groupby(['CIE', 'CNPJ'])['Avaliação'].mean().reset_index()
```

```python
reader = Reader(rating_scale=(0, 5))
data = Dataset.load_from_df(df_agrupado[['CIE', 'CNPJ', 'Avaliação']], reader)
```

Estamos criando nosso modelo de recomendação utilizando SVD. O SVD divide essa matriz de utilidade em três outras matrizes: U (matriz de usuários), Σ (matriz diagonal de valores singulares), e V^T (a transposta da matriz de itens).

Para prever a avaliação de um usuário para um item que ele ainda não avaliou, fazemos isso multiplicando a linha correspondente do usuário (U) a matriz de valores singulares truncados (Σ) e a coluna correspondente do item (V^T).

As métricas que escolhemos foram RMSE e MAE.

RMSE (Root Mean Square Error) e MAE (Mean Absolute Error) são duas métricas usadas para medir quão bem um modelo está prevendo algo comparado aos valores reais.

O RMSE é como uma média dos erros ao quadrado, dando mais peso a erros maiores. É como calcular a raiz quadrada da média dos erros ao quadrado. Se você tiver um grande erro, ele vai afetar mais o RMSE do que o MAE.

O MAE é uma média simples dos erros absolutos, sem elevar ao quadrado. Ele trata todos os erros, grandes ou pequenos, da mesma forma.

 Bem, essas métricas ajudam a avaliar o quão preciso é o sistema ao prever preferências do usuário. Em um sistema de recomendação, você quer que as sugestões sejam o mais próximas possível das preferências reais dos usuários.
 Com essas métricas conseguimos analisar a "gravidade" dos erros do sistema de recomendação.

Também é importante mencionar que dividimos nosso dataset em 90% para treino e 10% para teste.

```python
model = SVD()

results = cross_validate(model, data, measures=['RMSE', 'MAE'], cv=3, verbose=True)


trainset, testset = train_test_split(data, test_size=0.1)
```

No geral, com base nessas métricas, parece que o modelo está performando de maneira bastante consistente em todos os folds. Embora o MAE esteja em 1.4 (relativamente alto, já que uma nota realmente 3 pode variar entre 1,6 e 4,4), o RMSE está em apenas 1.7, nos dando a entender que embora ele erre bastante, esses erros não são grosseiros.

AGORA VAMOS TESTAR UTILIZANDO NOSSO DF DE TESTES. PARA FAZER A COMPARAÇÃO, COLOCAMOS A COLUNA DE AVALIAÇÃO REAL (COMO SE FOSSE A RESPOSTA CORRETA) E LOGO EM SEGUIDA COLOCAMOS A COLUNA DE AVALIAÇÃO PREVISTA(GERADA PELO MODELO)

```python
previsoes = []

for rating in testset:
    escola = rating[0]  # CIE Escola
    fornecedor = rating[1]  # CNPJ Fornecedor
    avaliacao_real = rating[2]  # Avaliação real

    # Fazendo previsão
    previsao = model.predict(escola, fornecedor)

    avaliacao_prevista = previsao.est

    previsoes.append({'Escola': escola, 'Fornecedor': fornecedor, 'Avaliação Real': avaliacao_real, 'Avaliação Prevista': avaliacao_prevista})

#Convertendo em dataframe
df_previsoes = pd.DataFrame(previsoes)

df_previsoes
```

Podemos considerar que os resultados parecem promissores, já que a coluna de avaliação prevista está bem próxima da avaliação real

Aqui estamos criando uma função que pega o CIE de uma escola, o CNPJ de uma fornecedora é aplica no modelo. O resultado é a nota prevista que aquela escola daria para o fornecedor.

```python
def fazer_previsao(CIE_escola, CNPJ_fornecedor):
    previsao = model.predict(CIE_escola, CNPJ_fornecedor)
    return previsao.est


fazer_previsao('122245','13.533.610/0001-00') # Previsão de uma escola que ainda não teve entrega com certa fornecedora
# Pelo resultado podemos analisar que a nota que ela daria para essa fornecedora seria essa abaixo.
```

Aqui criamos uma função que pega o CIE de uma escola e faz a previsão daquela escola com todos fornecedores disponíveis e retornar a nota prevista o nome daquele fornecedor que teria a maior avaliação.

```python
def melhor_fornecedora(CIE):
    df_fornecedoras_unicas = df_agrupado.drop_duplicates(subset='CNPJ')

    maior_valor = float('-inf')  # Inicializa com o menor valor possível
    nome_fornecedora_maior_valor = None

    for fornecedora in df_fornecedoras_unicas['CNPJ']:
        resultado_previsao = fazer_previsao(CIE, fornecedora)

        if resultado_previsao > maior_valor:
            maior_valor = resultado_previsao
            nome_fornecedora_maior_valor = fornecedora

    return nome_fornecedora_maior_valor, maior_valor

nome_fornecedora, maior_valor = melhor_fornecedora('12')
print(f'Maior valor retornado pela previsão: {maior_valor} para a fornecedora {nome_fornecedora}')

```

Podemos observar que o resultado para a escola de CIE 12 daria nota 4.61 para a fornecedora de CNPJ "71.615.942/0001-22". Sendo assim essa seria a ideal, já que essa nota foi comparada com todos fornecedores existentes e essa fornecedora teve maior nota.
